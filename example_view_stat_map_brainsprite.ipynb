```python
# one motor contrast map from NeuroVault
from nilearn import datasets
motor_images = datasets.fetch_neurovault_motor_task()
stat_img = motor_images.images[0]
```

```python
from nilearn.plotting import view_stat_map
view_stat_map(stat_img,threshold=2,cut_coords=[-38,-24,60],opacity=0.5,draw_cross=True,dim=0,colorbar=True,annotate=True,title="Motor Activation")
```
